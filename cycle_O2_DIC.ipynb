#### mask

```python
maskdir = '/scratch/usr/shklvn09/NUDATA/ORCA/05-data/meshmask/mesh_mask.nc'
tmask = xr.open_dataset(maskdir,
                      chunks={"z":46, "y":100, "x":100}).tmask.squeeze()
e1t = xr.open_dataset(maskdir,
                      chunks={"z":46, "y":100, "x":100}).e1t.squeeze()
e2t = xr.open_dataset(maskdir,
                      chunks={"z":46, "y":100, "x":100}).e2t.squeeze()
e3t = xr.open_dataset(maskdir,
                      chunks={"z":46, "y":100, "x":100}).e3t.squeeze()
```

```python
tmask, e3t = tmask.rename({'z':'deptht'}), e3t.rename({'z':'deptht'}) #renaming it for mask and tracers to have same dimension name
```

#### Computing global integral

LP01

```python
O2_int_01 = (O2_lp01_01.where(tmask) * e1t * e2t * e3t).sum(('x','y','deptht')) # applying mask to e1t, e2t, e3t gives the same results as computed here 
O2_int_02 = (O2_lp01_02.where(tmask) * e1t * e2t * e3t).sum(('x','y','deptht'))
O2_int_03 = (O2_lp01_03.where(tmask) * e1t * e2t * e3t).sum(('x','y','deptht'))
O2_int_04 = (O2_lp01_04.where(tmask) * e1t * e2t * e3t).sum(('x','y','deptht'))
O2_int_05 = (O2_lp01_05.where(tmask) * e1t * e2t * e3t).sum(('x','y','deptht'))
```

```python
O2_int_conc = np.hstack((np.array(O2_int_01.values).squeeze(), 
                       np.array(O2_int_02.values).squeeze(), 
                       np.array(O2_int_03.values).squeeze(),
                       np.array(O2_int_04.values).squeeze(),
                       np.array(O2_int_05.values).squeeze())) #concatenating all the cycles
```

```python
DIC_int_01 = (DIC_lp01_01.where(tmask) * e1t * e2t * e3t).sum(('x','y','deptht'))
DIC_int_02 = (DIC_lp01_02.where(tmask) * e1t * e2t * e3t).sum(('x','y','deptht'))
DIC_int_03 = (DIC_lp01_03.where(tmask) * e1t * e2t * e3t).sum(('x','y','deptht'))
DIC_int_04 = (DIC_lp01_04.where(tmask) * e1t * e2t * e3t).sum(('x','y','deptht'))
DIC_int_05 = (DIC_lp01_05.where(tmask) * e1t * e2t * e3t).sum(('x','y','deptht'))
```

```python
DIC_int_conc = np.hstack((np.array(DIC_int_01.values), 
                       np.array(DIC_int_02.values), 
                       np.array(DIC_int_03.values),
                       np.array(DIC_int_04.values),
                       np.array(DIC_int_05.values))) #concatenating all the cycles
```

```python
DICP_int_01 = (DICP_lp01_01.where(tmask) * e1t * e2t * e3t).sum(('x','y','deptht'))
DICP_int_02 = (DICP_lp01_02.where(tmask) * e1t * e2t * e3t).sum(('x','y','deptht'))
DICP_int_03 = (DICP_lp01_03.where(tmask) * e1t * e2t * e3t).sum(('x','y','deptht'))
DICP_int_04 = (DICP_lp01_04.where(tmask) * e1t * e2t * e3t).sum(('x','y','deptht'))
DICP_int_05 = (DICP_lp01_05.where(tmask) * e1t * e2t * e3t).sum(('x','y','deptht'))
```

```python
DICP_int_conc = np.hstack((np.array(DICP_int_01.values), 
                       np.array(DICP_int_02.values), 
                       np.array(DICP_int_03.values),
                       np.array(DICP_int_04.values),
                       np.array(DICP_int_05.values))) #concatenating all the cycles
```
